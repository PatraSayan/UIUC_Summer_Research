```python
import mdtraj as md
import numpy as np
```

```python
pdb_file = md.load('pentapeptide-impl-solv.pdb')

actual_trajectory = np.load('backup_train_test/y_test.pkl.npy', mmap_mode = None, allow_pickle = False, fix_imports=True,encoding='ASCII')
Predicted_trajectory = np.load('backup_train_test/Test_predict.pkl.npy', mmap_mode = None, allow_pickle = False, fix_imports = True, encoding = 'ASCII')

topology = pdb_file.topology
```

```python
actual_trajectory.shape
```

```python
Predicted_trajectory.shape
```

```python
def traject(n):
    new_pdb = pdb_file
    new_pdb.xyz = actual_trajectory[n]
    for x in new_pdb.topology.bonds:
        for y in topology.bonds:
            x = y
    new_pdb.save(f"PDB/label_{n}.pdb")
    
    new_pdb = pdb_file
    new_pdb.xyz = Predicted_trajectory[n]
    for x in new_pdb.topology.bonds:
        for y in topology.bonds:
            x = y
    new_pdb.save(f"PDB/predicted_{n}.pdb")
```
