```python
import mdtraj as md
import numpy as np
import pickle
```

```python
pdb_file = md.load('direct_pdbs/Actual_first/pentapeptide-impl-solv.pdb')


actual_trajectory = np.load('predictions and actual pkl files/5001_test_labels_with_hydrogen.pkl.npy', mmap_mode = None, allow_pickle = False, fix_imports=True,encoding='ASCII')
Predicted_trajectory = np.load('predictions and actual pkl files/5001_test_prediction_with_hydrogen.pkl.npy', mmap_mode = None, allow_pickle = False, fix_imports = True, encoding = 'ASCII')

topology = pdb_file.topology
```

```python
def actual_pdb(n):
    new_pdb = pdb_file
    new_pdb.xyz = actual_trajectory[n]
    for x in new_pdb.topology.bonds:
        for y in topology.bonds:
            x = y
    new_pdb.save("direct_pdbs/just_trajectory_change/09-07-2024/label_2826_frame.pdb")

def predicted_pdb(n):
    new_pdb = pdb_file
    new_pdb.xyz = Predicted_trajectory[n]
    for x in new_pdb.topology.bonds:
        for y in topology.bonds:
            x = y
    new_pdb.save("direct_pdbs/just_trajectory_change/09-07-2024/predicted_2826_frame.pdb")
```

```python
actual_pdb(2826)
predicted_pdb(2826)
```
